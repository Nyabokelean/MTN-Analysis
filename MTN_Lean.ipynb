# **ANALYSIS**

# Merging Datasets


```
# merging the three telcom datasets
tel_Combined = pd.concat([tel1, tel2, tel3], ignore_index=True)

tel_Combined
```

```
# deleting country A and country B column
del tel_Combined['COUNTRY_A']
tel_Combined.head(0)
```

```
# merging the telcom datasets with the cells dataset on the Site ID column
tel_all = tel_Combined.merge(cells, how='inner', on='SITE_ID')
tel_all.tail()
```

```
# Calculating the total number of voice calls made, messages sent and data used
tel_all.VALUE.groupby(tel_Combined['PRODUCT']).sum()
```

```
# Which time had alot of activities

group = tel_all.groupby(['PRODUCT', 'DATE_TIME'])['VALUE'].median()
group
```

```
# Calcutating the number of cities in the localizations(Abidjan & Interieur) in the cells geo dataset
cells.groupby('LOCALISATION').count()
```

```
# The most used cities for the three days
most = cells['VILLES'].value_counts()
most.head()
```
